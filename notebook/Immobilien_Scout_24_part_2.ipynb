## The dataset:
The datasets contains information about rental apartments in Berlin.
One row is for one appartment. The dataset contains 6 columns.  
- Region: is the Berlin district, where the flat is located
- Condition: is the condition of the flat
- Rooms: The number of rooms the flat has
- Rent: Monthly rent for the flat
- Year_Construction: The year in which the house was build
- Space: How many square meters does the flat have?

---
## The Task:
For each question, write code which clearly show the answer to it.

```python
# Imports
import pandas as pd
```

```python
# Read in dataset
rentals = pd.read_csv("https://raw.githubusercontent.com/juliandnl/redi_ss20/master/berlin_rental.csv")
rentals.head()
```

1.How many apartments in Mitte are in a 'well-kept' condition?

```python
rentals.loc[rentals["Condition"] == "well_kept","Condition"].value_counts()
```

```python
rentals.loc[rentals["Region"] == "Mitte" ,"Condition"]
```

2.How many apartments in Kreuzberg were built before 1900?

```python
r = rentals.loc[rentals["Year_Construction"] < 1900,"Year_Construction"]
len(r)
```

3.What are the average apartment sizes by the number of rooms?

```python
rentals.groupby("Rooms")["Space"].mean().round(2)
```

4.Which is the most expensive region on average and what is the maximum rent for this region?

```python
rentals.groupby("Region")["Rent"].agg(["mean","max"]).sort_values(by=['mean'],ascending=False).reset_index().head(1)
```

## * Bonus Question:
5. What are the minimum, maximum and average number of rooms of the apartments in Köpenick?


```python
rentals.groupby("Region")["Rooms"].agg(["min","max","mean"]).loc["Köpenick"]
```

## ** Extra Bonus Question
6. Show a table of apartment counts, where rows and columns correspond to year of construction and region, respectively.

```python
year_and_region = rentals.groupby(["Year_Construction","Region"])[["Condition"]].count()
year_and_region.rename(columns={"Condition":"Apartments"}, inplace=True)
year_and_region_wide = pd.pivot_table(year_and_region,values="Apartments",index="Year_Construction",columns="Region")
year_and_region_wide
```
